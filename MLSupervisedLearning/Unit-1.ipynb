## Distance metrics

`euclidean`
$$\sqrt{\sum_i(x_i-y_i)^2}$$


`cityblock` (Manhattan)
$$\sum_i(|x_i-y_i|)$$

`minkowski`
- args `p`
$$(\sum_i (|x_i - y_i|^p )^{1/p}$$

`mahalanobis`
- args `VI` inverse of the covariance matrix
$$\sqrt{(x - y)^T V^{-1} (x-y)}$$


`hamming`
$$\frac{\sum_i I(x_i \neq y_i)}{d}$$

`cosine` (distance)
$$1 - \frac{x \cdot y}{\lVert x \rVert_2 \cdot \lVert y \rVert_2}$$

`correlation` (distance)
$$1 - \frac{(x - \bar{x}) \cdot (y - \bar{y})}{\lVert x - \bar{x} \rVert_2 \cdot \lVert y - \bar{y} \rVert_2}$$


`jaccard`
$$\frac{\sum_i I(x_i \neq y_i \land x_i > 0 \land y_i > 0)}{\sum_i I(x_i > 0 \lor y_i > 0)}$$


https://docs.scipy.org/doc/scipy/reference/spatial.distance.html

```python
import numpy as np
```

```python
from scipy.spatial.distance import pdist, squareform
```

### Simple example

- $x = (0, 3)$
- $y = (4, 0)$
- $V^T = 
\begin{bmatrix}
    1 & 0 \\
    0 & 1
\end{bmatrix}
$ (inverse covariance matrix)

```python
x = np.array([0,3])
y = np.array([4,0])
VI = np.array([[1,0],[0,1]]) # inverse covariance matrix
```

```python
X = np.stack([x,y], axis=0)
X
```

```python
pdist(X, metric='euclidean')
```

```python
pdist(X, metric='cityblock')
```

```python
pdist(X, metric='minkowski', p=1)
```

```python
pdist(X, metric='minkowski', p=2)
```

```python
pdist(X, metric='mahalanobis', VI=VI)
```

```python
pdist(X, metric='hamming') # normalized by the number of dimensions
```

```python
pdist(X, metric='cosine')
```

```python
pdist(X, metric='correlation')
```

```python
pdist(X, metric='jaccard')
```

```python
x = np.array([38, 1, 6, 10, 20, 100000])
y = np.array([35, 1, 6, 10, 10, 60000])
X = np.stack([x,y], axis=0)
for metric in ['hamming', 'cosine', 'jaccard']:
    dist = pdist(X, metric=metric)
    print('{}: {}'.format(metric, dist))
```

## Heatmap plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
metrics = {'euclidean':[{}],
           'cityblock':[{}],
           'minkowski':[{'p':1}, {'p':2}],
           'hamming':[{}],
           'cosine':[{}],
           'correlation':[{}],
           'jaccard':[{}]}
```

```python
fig, axes = plt.subplots(nrows=8, ncols=1, figsize=(10, 80))


img = 0
for metric in metrics:
    for kwargs in metrics[metric]:
        dist = pd.DataFrame(squareform(pdist(transaction_df, metric=metric, **kwargs)),
                    index=transaction_df.index,
                    columns=transaction_df.index)
#         print(dist)

        sns.heatmap(dist, ax=axes[img])
        axes[img].set_title('{}:{}'.format(metric, kwargs))
        img += 1
        
        
```
